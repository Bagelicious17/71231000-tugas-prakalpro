```python
%matplotlib inline

import matplotlib.pyplot as plt 
import numpy as np

x=np.linspace(0,10)
y=np.sin(x)
z=np.cos(x)

plt.plot(x,y,'b',x,z,'r')
plt.xlabel('radians');
plt.ylabel('value');
plt.title('plotting demonstration')
plt.legend(['Sin','Cos'])
plt.grid()
```

```python
p=0
while p < 2 :
    p+=1
    jumlah_gram=int(input("masukkan jumlah gram emas:"))
    harga=int(input("harga emas yang dibeli gerard:"))
    harga_akhir=int(input("harga emas setelah mengalami kenaikan:"))
    persenan=float((harga_akhir-harga)/harga*100)
    profit= (harga_akhir-harga)*jumlah_gram
    print(f"total keuntungan gerard dalam rp adalah Rp.{profit}")
    print(f"total keuntungan gerard dalam persenan adalah {persenan}%")
```

```python
uang=float(input("masukkan uang awal:"))
uang_akhir=int(input("masukkan uang tujuan:"))
bunga=float(input("masukkan jumlah bunga per tahun:"))
s=0
while uang < uang_akhir:
    t=float(uang*bunga/100)
    uang+=t
    s+=1
print(f"jumlah tahun yang diperlukan adalah {s}, untuk memperoleh jumlah uang {uang}.")
```
